```
import pandas as pd
import pyspark

#Ablageort
# dbfs:/FileStore/tables/accountholder_csv_00000_of_00001
# dbfs:/FileStore/tables/account_csv_00000_of_00001-1
# dbfs:/FileStore/tables/product_csv_00000_of_00001
# dbfs:/FileStore/tables/transaction_csv_00000_of_00001

#Import
accountholder_df1 = spark.read.format("csv").option("escape", "\"").load("dbfs:/FileStore/tables/accountholder_csv_00000_of_00001")
account_df1 = spark.read.format("csv").option("escape", "\"").load("dbfs:/FileStore/tables/account_csv_00000_of_00001-1")
product_df1 = spark.read.format("csv").option("escape", "\"").load("dbfs:/FileStore/tables/product_csv_00000_of_00001")
transaction_df1 = spark.read.format("csv").option("escape", "\"").load("dbfs:/FileStore/tables/transaction_csv_00000_of_00001-1")

display(transaction_df1)
```

```
import pandas as pd

account_df = account_df1.select("*").toPandas()
accountholder_df = accountholder_df1.select("*").toPandas()
product_df = product_df1.select("*").toPandas()
transaction_df = transaction_df1.select("*").toPandas()

transaction_df
```

```
#######Rename Column#########
account_df.columns = [str(column) for column in account_df.columns]
account_df = account_df.rename(columns={'_c0': 'account_id', '_c1': 'account_holder_id', '_c2': 'account_holder_name', '_c3': 'iban', '_c4': 'bic', '_c5': 'balance', '_c6': 'currency_code', '_c7': 'product_id'})
#Drop Column "Bic", because of no values
account_df = account_df.dropna(how='all', axis=1)


product_df.columns = [str(column) for column in product_df.columns]
product_df = product_df.rename(columns={'_c0': 'product_id', '_c1': 'product_name', '_c2': 'product_desc', '_c3': 'product_type'})

accountholder_df.columns = [str(column) for column in accountholder_df.columns]
accountholder_df = accountholder_df.rename(columns={'_c0': 'account_holder_id', '_c1': 'account_holder_type', '_c2': 'title', '_c3': 'first_name', '_c4': 'last_name', '_c5': 'birth_name', '_c6': 'date_of_birth', '_c7': 'place_of_birth', '_c8': 'phone_number', '_c9': 'email', '_c10': 'country', '_c11': 'zip_code', '_c12': 'city', '_c13': 'street', '_c14': 'house_number', '_c15': 'address_type'})
   
transaction_df.columns = [str(column) for column in transaction_df.columns]
transaction_df = transaction_df.rename(columns={'_c0': 'account_id', '_c1': 'transaction_id', '_c2': 'reason_for_payment', '_c3': 'transaction_text', '_c4': 'amount', '_c5': 'number_of_decimals', '_c6': 'direction', '_c7': 'currency_code', '_c8': 'booking_status', '_c9': 'booking_date', '_c10': 'value_date', '_c11': 'is_returnable'})



account_df
```

```
%sql
CREATE DATABASE Bank
```

```
sparkaccount_df = spark.createDataFrame(account_df)
sparkaccountholder_df = spark.createDataFrame(accountholder_df)
sparkproduct_df = spark.createDataFrame(product_df)
sparktransaction_df = spark.createDataFrame(transaction_df)

```

```
sparkaccountholder_df.write.mode("overwrite").saveAsTable("bank.AccountHolder")
sparkaccount_df.write.mode("overwrite").saveAsTable("bank.Account")
sparkproduct_df.write.mode("overwrite").saveAsTable("bank.Product")
sparktransaction_df.write.mode("overwrite").saveAsTable("bank.TransactionX1")
```

```
%sql
SELECT bank.Account.account_id as Account_ID,bank.Account.account_holder_name,  SUM (bank.TransactionX1.amount) as sum_amount 
FROM bank.Account
left join bank.TransactionX1 on bank.Account.account_id=bank.TransactionX1.account_id
GROUP BY bank.Account.account_id, bank.Account.account_holder_name;

```

```
# DDL Aufbau fr Use-Case
import sqlite3
conn = sqlite3.connect('bank.db')
# Create Cursor
c = conn.cursor()
c.execute("DROP TABLE IF EXISTS Account")
c.execute("DROP TABLE IF EXISTS AccountHolder")
c.execute("DROP TABLE IF EXISTS TransactionX")
c.execute("DROP TABLE IF EXISTS Product")

# Create Account Table
c.execute(""" CREATE TABLE Account (
    account_id INT PRIMARY KEY,
    account_holder_id TEXT,
    account_holder_name TEXT,
    iban TEXT,
    bic TEXT,
    balance REAL,
    currency_code TEXT,
    product_id INT
   
)""")
# FOREIGN KEY(account_holder_id) REFERENCES AccountHolder(account_holder_id)
# FOREIGN KEY(product_id) REFERENCES Product(product_id)
c.execute(""" CREATE TABLE Product (
    product_id INT PRIMARY KEY,
    product_name TEXT,
    product_desc TEXT,
    product_type INT
    
)""")
#FOREIGN KEY(product_id) REFERENCES Account(product_id)
c.execute(""" CREATE TABLE AccountHolder (
    account_holder_id INT PRIMARY KEY,
    account_holder_type TEXT,
    title TEXT,
    first_name TEXT,
    last_name TEXT,
    birth_name TEXT,
    date_of_birth TEXT,
    place_of_birth TEXT,
    phone_number TEXT,
    email TEXT,
    country TEXT,
    zip_code INT,
    city TEXT,
    street TEXT,
    house_number INT,
    address_type TEXT
    
)""")
#FOREIGN KEY(account_holder_id) REFERENCES Account(account_holder_id)
c.execute(""" CREATE TABLE TransactionX (
    account_id INT,
    transaction_id INT PRIMARY KEY,
    reason_for_payment TEXT,
    transaction_text TEXT,
    amount REAL,
    number_of_decimals INT,
    direction TEXT,
    currency_code TEXT,
    booking_status TEXT,
    booking_date TEXT,
    value_date TEXT,
    is_returnable TEXT
    
)""")
#FOREIGN KEY(account_id) REFERENCES Account (account_id)

#Commit
conn.commit()

#Close connection
# conn.close()
```
